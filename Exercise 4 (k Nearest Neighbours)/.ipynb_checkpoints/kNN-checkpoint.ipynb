# k-Nearest Neighbor

```python
# YOUR CODE HERE
import math

labels = {}

def classification (input_values):
  cl_y_test = []
  
  distances = {}

  for coordinate in input_values:
    print (coordinate)
    coordinate_value = coordinate[0]
    print("From " + str(coordinate) + str(", the distances are:"))
    for point in range(len(cl_X)):
      squared_distance = 0
      labels[cl_X[point][0]] = cl_y[point]

      squared_distance += (coordinate_value - cl_X[point]) ** 2
      distances[cl_X[point][0]] = (math.sqrt(squared_distance))
    print ("Distances:", distances, "\n")
    print ("Labels:", labels, "\n")

    sorted_keys = [key for key, _ in sorted(distances.items(), key=lambda item: item[1])]

    print("Points in order of distance from " + str(coordinate) + ":", sorted_keys, "\n")

    k = int(input("Input k: "))

    shortlisted_points = list(sorted_keys)[:k]

    print("k points based on distance:", shortlisted_points, '\n')

    scores = {}

    for element in shortlisted_points:
      # print (element)
      label = labels[element]
      if label not in scores:
        scores[label] = 1
      else:
        scores[label] += 1

    print("Scores:", scores, '\n')

    num_of_scores = len(scores)

    first_key = list(scores.keys())[:1]
    first_value = scores[first_key[0]]
    if num_of_scores > 1:
      first_key, second_key = list(scores.keys())[:2]
      second_value = scores[second_key]
      if first_value > second_value:
          print(f"The score of label {first_key} is the highest. Hence, the label of", coordinate_value, "is also", first_key)
          cl_y_test.append(first_key)
      elif first_value == second_value:
          print(f"The score of label {first_key} is equal to the score of label {second_key}. Hence, the label of", coordinate_value, "can be either", first_key, "or", second_key, "\n")
      if num_of_scores > 2:
        first_key, second_key, third_key = list(scores.keys())[:3]
        third_value = scores[third_key]
        if first_value == second_value and second_value == third_value:
              print ("More than 2 answers possible. Choose another value of k.")
    else:
      print(f"Since {first_key[0]} is the only label, the label of", coordinate_value, "is also", first_key[0])
      cl_y_test.append(first_key[0])
  return("cl_y_test =", cl_y_test)

def regression (input_values):

  r_y_test = []
  distances = {}
  for coordinate in input_values:
    print("From " + str(tuple(coordinate)) + str(", the distances are:"))
    for point in range(len(r_X)):
      squared_distance = 0
      labels[tuple(r_X[point])] = r_y[point]
      for axis in range(len(coordinate)):
        squared_distance += (coordinate[axis] - r_X[point][axis]) ** 2
      distances[tuple(r_X[point])] = (math.sqrt(squared_distance))
    print ("Distances:", distances, "\n")
    # print ("Labels:", labels, "\n")
    sorted_keys = [key for key, _ in sorted(distances.items(), key=lambda item: item[1])]

    print("Points in order of distance from " + str(coordinate) + ":", sorted_keys, "\n")

    k = int(input("Input k: "))

    shortlisted_points = list(sorted_keys)[:k]

    print("k points based on distance:", shortlisted_points, '\n')

    sum_of_labels = 0
    for neighbour in shortlisted_points:
      # print (labels[neighbour])
      sum_of_labels += labels[neighbour]
    mean = sum_of_labels / k

    print ("Based on k = " + str(k) + ", the label of", coordinate, "is", mean)
    r_y_test.append(mean)
  return ("r_y_test =", r_y_test)
```

## Compare the results with the sklearn kNN algorithm

```python
# Make your code work with the following inputs:
# Use this data as your reference data points and labels

# Classification data
cl_X = np.arange(0, 9).reshape(9,1)
cl_y = np.array([0, 0, 0, 1, 1, 1, 2, 2, 2])
# print('Classification input:\n', cl_X, '\tShape:', cl_X.shape)
# print('Classification labels:\n', cl_y, '\tShape:', cl_y.shape)


# Regression data
r_X = np.array([[1, 2], [2, 4], [3, 6], [4, 8], [5, 10]])
r_y = np.array([1, 2, 3, 4, 5])
# print('Regression input:\n', r_X, '\tShape:', r_X.shape)
# print('Regression labels:\n', r_y, '\tShape:', r_y.shape)


# Use the following list to test your code for classification
cl_X_test = np.array([[2.1], [5.2], [7.2]])

r_X_test = np.array([[2.1, 5.1], [2.6, 6.2]])
```

```python
# classification (dont forget the weighting function)

# yours
print(classification(cl_X_test))

# sklearn comparison
from sklearn.neighbors import KNeighborsClassifier

# Create Classifier
knn_clf = KNeighborsClassifier(n_neighbors=3) # Use 3 as a placeholder for k
knn_clf.fit(cl_X, cl_y)

# Test the classifier
predicted_sklearn_classifier = knn_clf.predict(cl_X_test)

print (predicted_sklearn_classifier)
```

```python
# regression (dont forget the weighting function)

# yours
print(regression(r_X_test))

# sklearn comparison
from sklearn.neighbors import KNeighborsRegressor

# Create Regressor
knn_reg = KNeighborsRegressor(n_neighbors=3) # Similarly, briefly assuming k=3
knn_reg.fit(r_X, r_y)

# Test the regressor
predicted_sklearn_regressor = knn_reg.predict(r_X_test)
print (predicted_sklearn_regressor)
```
